```python
import os
import platform
import matplotlib.font_manager as font_manager
import matplotlib.pyplot as plt
import pandas as pd
from typing import Optional
import logging as lg

def set_plt() -> None:
    ## set plt minus font
    plt.rcParams["axes.unicode_minus"] = False

    ## set font
    system_name = platform.system()
    if system_name == "Windows":
        # Windows
        plt.rc("font", family="Malgun Gothic")
    elif system_name == "Darwin":
        # Mac
        plt.rc("font", family="AppleGothic")
    elif system_name == "Linux":
        # Linux
        path = "/usr/share/fonts/truetype/nanum/NanumMyeongjo.ttf"
        font_name = font_manager.FontProperties(fname=path, size=12)
        plt.rc("font", family=font_name)
    else:
        print("Not support")

set_plt()
df = pd.read_csv("files/노인의_우울증상.csv", encoding="euc-kr",header=[0,1])
df
```

## 성별 우울증 현황

```python
x_label = ['2011','2014','2017','2020']

# 남자
m1 = df.loc[3,('2011','우울증상 (%)')].copy()
m2 = df.loc[3,('2014','우울증상 (%)')].copy()
m3 = df.loc[3,('2017','우울증상 (%)')].copy()
m4 = df.loc[3,('2020','우울증상 (%)')].copy()

# 여자
f1 = df.loc[4,('2011','우울증상 (%)')].copy()
f2 = df.loc[4,('2014','우울증상 (%)')].copy()
f3 = df.loc[4,('2017','우울증상 (%)')].copy()
f4 = df.loc[4,('2020','우울증상 (%)')].copy()

m_y = [m1, m2, m3, m4]
f_y = [f1, f2, f3, f4]

plt.title('성별 우울증 변화 그래프')
plt.plot(x_label, m_y, 'skyblue', label='남자')
plt.plot(x_label, f_y, 'hotpink', label='여자')
plt.ylabel('우울증상 (%)')
plt.legend()
plt.show()
```
