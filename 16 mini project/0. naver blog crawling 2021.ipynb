# 블로그 내용 크롤링

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
import datetime
import time
```

```python
driver = webdriver.Chrome()
```

```python
blog_body = pd.DataFrame(columns = ["body", "label"])
```

```python
url = "https://search.naver.com/search.naver?where=blog&query=%EA%B3%BC%ED%95%99&sm=tab_opt&nso=so:r,p:from20180101to20210101"

driver.get(url)

scroll_location = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(1)
    scroll_height = driver.execute_script("return document.body.scrollHeight")

    if scroll_location == scroll_height:
        break
    else:
        scroll_location = driver.execute_script("return document.body.scrollHeight")
    
element = driver.find_elements_by_class_name("api_txt_lines")

to_str = []
to_df = pd.DataFrame(columns = ["body"])

for i in list(range(len(element))):
    to_str.append(element[i].text)
    
to_df = pd.DataFrame(to_str)
blog_body["body"] = to_df

blog_body
```

```python
blog_body["label"] = 1
blog_body
```

```python
blog_body = blog_body.drop('label',axis=1)
```

```python
blog_body.to_csv("blog_2019_16.csv", encoding ="utf-16", index = False)
```
